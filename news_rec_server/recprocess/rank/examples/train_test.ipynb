```python
import pandas as pd
import matplotlib.pyplot as plt
df_1 = pd.read_csv('deepfm.csv')
df_1['epoch'] = range(1, len(df_1) + 1)
df_2 = pd.read_csv('deepfm_ppnet.csv')
df_2['epoch'] = range(1, len(df_2) + 1)
df = pd.merge(left=df_1, right=df_2,how='outer', on='epoch', suffixes=('_deepfm', '_ppnet'))
df.rename(columns={'loss_deepfm':'deepfm_train_loss', 'auc_deepfm':'deepfm_train_auc', 'val_loss_deepfm':'deepfm_val_loss', 'val_auc_deepfm':'deepfm_val_auc',
                   'loss_ppnet':'ppnet_train_loss', 'auc_ppnet':'ppnet_train_auc', 'val_loss_ppnet':'ppnet_val_loss', 'val_auc_ppnet':'ppnet_val_auc',}, inplace=True)
```

```python
# 创建图形和轴对象
fig, ax = plt.subplots()

# 绘制折线图
ax.plot(df['epoch'], df['deepfm_train_loss'], label='DeepFM Train Loss')
ax.plot(df['epoch'], df['ppnet_train_loss'], label='PPNet Train Loss')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')

# 显示图形
plt.show()

# 保存图形到文件
fig.savefig('train_loss.svg',format='svg', dpi=150, pad_inches=0.1)

# 关闭图形，释放内存
plt.close(fig)
```

```python
# 创建图形和轴对象
fig, ax = plt.subplots()

# 绘制折线图
ax.plot(df['epoch'], df['deepfm_train_auc'], label='DeepFM Train Auc')
ax.plot(df['epoch'], df['ppnet_train_auc'], label='PPNet Train Auc')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('Auc')

# 显示图形
plt.show()

# 保存图形到文件
fig.savefig('train_auc.svg',format='svg', dpi=150, pad_inches=0.1)

# 关闭图形，释放内存
plt.close(fig)
```

```python
# 创建图形和轴对象
fig, ax = plt.subplots()

# 绘制折线图
ax.plot(df['epoch'], df['deepfm_val_loss'], label='DeepFM Val Loss')
ax.plot(df['epoch'], df['ppnet_val_loss'], label='PPNet Val Loss')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')

# 显示图形
plt.show()

# 保存图形到文件
fig.savefig('val_loss.svg',format='svg', dpi=150, pad_inches=0.1)

# 关闭图形，释放内存
plt.close(fig)
```
